```python
import requests
```

```python
questions_url = 'https://oda.ft.dk/api/Dokument?$inlinecount=allpages&$filter=typeid%20eq%2016'
```

```python
#This function takes dict, extracts the next link, gets the new dict and appends it to the old dict


def get_all_questions(url):

    r = requests.get(url)

    questions = r.json()
    prev = r.json()
    page = 1

    while 'odata.nextLink' in prev:

        r = requests.get(prev['odata.nextLink'])
        questions['value'] += r.json()['value']
        print(f'Page {page} done')
        page += 1
        prev = r.json()


    return questions
```

```python
import regex
```

```python
#creating a regex pattern to identify the location of the sequence 'Svar: ' anywhere in the text
expression = '\w*(Svar: *)'
```

```python
#we convert to text using pymupdf
import fitz

def pdf_to_text(pdf_path):
    try:
        doc = fitz.open(pdf_path)
        text = ''
        for page in doc:
            text += page.get_text("text")
        
        #using regex to check if the word 'Svar: ' is in the text (regardless of what is in front or after) and if it is, we find the index of the word and slice the text from that index
        if regex.search(expression, text):
            index = regex.search(expression, text).start()
            text = text[index:]

    except:
        text = None

    return text

#We convert all the pdfs in the answers folder to text
import os

#find all the pdfs in the folder
pdfs = [i for i in os.listdir('answers') if i.endswith('.pdf')]


```

```python
#We convert all the pdfs to text
texts = [pdf_to_text(f'answers/{i}') for i in pdfs]
```
